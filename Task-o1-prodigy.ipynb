```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample dataset on ages and genders in a population
data = {
    "Age": [22, 25, 29, 35, 45, 55, 65, 70, 23, 31, 38, 42, 48, 54, 60, 68, 74, 28, 33, 36, 47, 52, 57, 62, 66],
    "Gender": ["Male", "Female", "Female", "Male", "Female", "Male", "Female", "Male", "Male", "Female", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female"]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Plotting the distribution of ages
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10, kde=True)
plt.title('Distribution of Ages in the Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
